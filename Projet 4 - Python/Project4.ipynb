### ANALYSE DES EMISSIONS DE CO2 PAR PAYS ENTRE 1960 ET 2019

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import du fichier CSV
df = pd.read_csv('./co2_emissions_kt_by_country.csv')
df
```

```python
df.describe() #résumé des données numériques
```

```python
df.isna().sum() #vérification des valeurs manquantes
```

```python
df.info() #vérification des formats de colonnes
```

##### TOTAL PAR AN

```python
df.groupby('year', axis=0).sum() #groupement des données par année
```

```python
sns.lineplot(x='year', y='value', data=df.groupby('year', axis=0).sum())
plt.show()
```

#### EMISSIONS TOTALES PAR PAYS

```python
# émissions totales de co2 par pays
df2 = df.groupby('country_name', axis=0).sum()
df2.drop(['country_code', 'year'], axis=1, inplace=True)
df2
```

#### EMISSIONS POUR UN PAYS

```python
# Sélectionner les émissions pour un pays
df.loc[df['country_name']=='France'].head()
```

```python
sns.lineplot(x='year', y='value', data=df.loc[df['country_name']=='France'])
plt.show()
```

#### COMPARATIF ENTRE PAYS

```python
# Comparaison entre plusieurs pays
sns.lineplot(x='year', y='value', data=df.loc[df['country_name']=='France'], color='blue', label='France')
sns.lineplot(x='year', y='value', data=df.loc[df['country_name']=='Italy'], color='red', label='Italy')
sns.lineplot(x='year', y='value', data=df.loc[df['country_name']=='Spain'], color='black', label='Spain')
plt.legend(loc='lower right')
plt.show()
```

```python
# Comparaison des émissions totales entre plusieurs pays
df3 = df.loc[(df['country_name']=='France') | (df['country_name']=='Italy') | (df['country_name']=='Spain')].groupby('country_name', axis=0).sum()
df3.drop(['country_code', 'year'], axis=1, inplace=True)
df3
```

```python
df3.plot.bar()
```

#### TOTAL PAR PAYS PAR VOLUME D'EMISSIONS

```python
# Pays les plus éméteurs
df2.sort_values(['value'], ascending=False, axis=0, inplace=True)
df2.head(21)
```

```python
df2.iloc[1:21].plot.bar()
```

```python
# Pays les moins éméteurs
df2.tail(21)
```

```python
df2.tail(21).sort_values(['value'], ascending=True, axis=0).plot.bar()
```
